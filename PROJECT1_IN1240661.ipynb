# Cleaning Data on copy of the dataset

```python
copy=pd.read_excel(file_path)
print(copy)
```

# Changing the DOL records where the value is 'Present' to today's date 

```python
today_date=pd.to_datetime("today").strftime("%Y-%m-%d")
#changing the present to todays date
copy['DOL']=copy['DOL'].apply(lambda x:today_date if x=='present' else x)
print(copy['DOL'])
```

# Convert into datetime object


```python
lst=['DOL','DOJ','DOB']
def changeType(x):
    copy[x]=pd.to_datetime(copy[x])
    return "Datatype changed"
for i in lst:
    print(changeType(i))
```

```python
copy.info()
```

# Remove the first unnamed column 

```python
copy=copy.loc[:,~copy.columns.str.contains('^Unnamed')]
print(copy)
```

# Finding Outliers in each numerical column

# Salary Column

```python
import matplotlib.pyplot as plt
salary_data=copy['Salary'].values.tolist()
plt.boxplot(salary_data)
plt.show()
```

```python
import numpy as np
q1=np.percentile(salary_data,25)
q3=np.percentile(salary_data,75)
iqr=q3-q1
lb=q1-(1.5)*iqr
ub=q3+(1.5)*iqr
after_removing=[]
for i in salary_data:
    if i>lb and i<ub:
        after_removing.append(i)
plt.boxplot(after_removing)
print("After removing the outliers")
plt.show()
```

# Bivariate Analysis

# Scatter plot of salary and collegeGpa

```python
plt.scatter(salary_data,clg_gpa_data)
plt.xlabel("Salaries(x 10000000)")
plt.ylabel("CollegeGPA")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
top_designations = copy['Designation'].value_counts().head(5)
mean_salaries = copy.groupby('Designation')['Salary'].mean().loc[top_designations.index]
fig, ax1 = plt.subplots()
color = 'blue'
ax1.set_xlabel('Designation')
ax1.set_ylabel('Occurrences', color=color)
ax1.bar(top_designations.index, top_designations, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='x',labelcolor=color,rotation=90)
ax2 = ax1.twinx()
color = 'red'
ax2.set_ylabel('Mean Salary', color=color)
ax2.plot(top_designations.index, mean_salaries, color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)
```
